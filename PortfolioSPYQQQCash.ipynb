```python
df['SMA_10_x'] = df.loc[:, 'Adj Close_x'].rolling(window=10).mean()
```

```python
def runMVA(portfolio, SellTolerance, BuyTolerance):
    #portfolio = df[['Date', 'Adj Close_x', 'SMA_10_x']]
    # resLabel='x'
    # runMVA(portfolio)
    
    numShares=np.zeros(portfolio.shape[0]) # shares held throughout just before month end buying/selling
    StartingBalance=10000
    newBal=np.zeros(portfolio.shape[0]) #Balance on month end just before buying/selling using AdjClosingPrice
    
    # This algo holds either shares or cash - simple algo sell all or hold all

    # Start from first row - each row is a month recorded at end of month Closing Price. NumShares reflects just before selling i.e. on 31st. and starting 1 if would be 0.
    mode=1 #1=buy, 0=sell/hold
    newBal[0] = StartingBalance

    for i in range(1, portfolio.shape[0]):
        if pd.isna(portfolio.iloc[i,2]): #dont start algo till u have 10months mva ready - TBD: backfill previous values
            newBal[i]=newBal[i-1] 
            numShares[i]=numShares[i-1] #assume numShares is 0 till we buy
            continue
        
        # what was our state prevEOM
        if numShares[i-1] > 0: #had shares just before end of last month
            if portfolio.iloc[i-1,1] / portfolio.iloc[i-1,2] < SellTolerance:
                #if own and price<mva then we sold on 1st
                numShares[i]=0
                newBal[i] = numShares[i-1] * portfolio.iloc[i-1,1] #we sold on end of 1st at end of last month's closing price. But since its cash, its same EOM.
            else:
                #maintain
                numShares[i]=numShares[i-1]
                newBal[i] = numShares[i] * portfolio.iloc[i,1] # holding shares so reflect current share price
        else:
            # Zero shares held. i.e. we were in Cash
            if portfolio.iloc[i-1,1] >= portfolio.iloc[i-1,2]: #It was buy signal prevEOM
                numShares[i] = newBal[i-1] / portfolio.iloc[i-1,1] # We bought on 1st using prev EOM CLosing Price.
                newBal[i] = numShares[i] * portfolio.iloc[i,1] # EOM bal reflects shares * priceClosingEOM
            else:
                numShares[i]=numShares[i-1] # its all zero. But in future may do other algos
                newBal[i] = newBal[i-1] # cash is unchanged

    return numShares, newBal
    
    
```

```python
#portfolio = df[['Date', 'Adj Close_x', 'SMA_10_x']]
#resLabel='x'
#numShares, newBal = runMVA(portfolio, 0, 1)
#df['numShares_'+resLabel] = numShares
#df['newBal_'+resLabel] = newBal


portfolio = df[['Date', 'Adj Close_y', 'SMA_10_y']]
resLabel='y'
numShares, newBal = runMVA(portfolio, 0.99, 1)
df['numShares_'+resLabel] = numShares
df['newBal_'+resLabel] = newBal
```

```python
def drawStrategy(data):
    #data = [Date, newBal, AdjClose, SMA_10]
    fig, ax1 = plt.subplots(figsize=[15,10])
    ax2 = ax1.twinx()
    plt.grid(True, 'both')
    x=data.iloc[:,0]
    l1= ax1.plot(x, data.iloc[:,1], 'b-')
    ax1.set_xticklabels(x, rotation=60)
    ax1.legend(loc='upper right', shadow=True)
    #l3= ax2.plot(x, df['Adj Close_x']/ (df['Adj Close_x'][0]), 'b.')
    
    l2= ax2.plot(x, data.iloc[:,2], 'r-.')
    l3= ax2.plot(x, data.iloc[:,3], 'g:')
    #l4= ax2.plot(x, df['Adj Close_y']/df['Adj Close_y'][0], 'm.')
    ax2.legend(loc=2)
    plt.show()
```

```python
SPYData = df[['Date', 'newBal_y', 'Adj Close_y', 'SMA_10_y', 'numShares_y']]
drawStrategy(SPYData)
```
